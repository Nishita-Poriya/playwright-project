# Playwright Alert Test Debug Runner

This notebook runs the fixed Playwright alert test file in headed mode and prints the results.

```python
import subprocess
from pathlib import Path

notebook_dir = Path(r"c:\Users\Welcome\Playwright\playwright-project")
command = ["npx", "playwright", "test", "alert.test.ts", "--headed", "--reporter=list"]
proc = subprocess.run(command, cwd=notebook_dir, capture_output=True, text=True)
print("returncode:", proc.returncode)
print("stdout:\n", proc.stdout)
print("stderr:\n", proc.stderr)
```
